# Thực hành numpy

```python
import numpy as np

# 1. Tạo mảng cơ bản
a = np.array([1, 2, 3])
b = np.array([(1.5, 2, 3), (4, 5, 6)], dtype=float)

# 2. Khởi tạo mảng đặc biệt
zeros_arr = np.zeros((3, 4)) # Mảng toàn số 0
ones_arr = np.ones((2, 3, 4)) # Mảng toàn số 1
arange_arr = np.arange(10, 25, 5) # Mảng từ 10 đến 25, bước nhảy 5
linspace_arr = np.linspace(0, 2, 9) # 9 số cách đều từ 0 đến 2

# 3. Các phép toán ma trận
print("Tổng mảng a:", a.sum())
print("Giá trị nhỏ nhất mảng b:", b.min())
print("Ma trận b sau khi chuyển vị:\n", b.T)
```

```python
import pandas as pd

# 1. Tạo DataFrame từ Dictionary
data = {
    'Country': ['Belgium', 'India', 'Brazil'],
    'Capital': ['Brussels', 'New Delhi', 'Brasília'],
    'Population': [11190846, 1303171035, 207847528]
}
df = pd.DataFrame(data, columns=['Country', 'Capital', 'Population'])

# 2. Lấy thông tin cơ bản
print("3 dòng đầu tiên:\n", df.head(3))
print("Kích thước bảng:", df.shape)
print("Thông tin chung:\n", df.info())

# 3. Truy xuất dữ liệu
# Lấy theo vị trí (Hàng 0, Cột 0)
print("Phần tử đầu tiên:", df.iloc[0, 0])
# Lấy theo tên (Dòng 0, Cột Capital)
print("Thủ đô dòng đầu:", df.loc[0, 'Capital'])
```

```python
# Numpy sang Pandas
h = np.array([[1, 2], [3, 4]])
df_h = pd.DataFrame(h)
print("DataFrame từ Numpy:\n", df_h)

# Pandas sang Numpy
numpy_h = df_h.to_numpy() # Cách này chuẩn hơn np.array(df_h) trong các bản mới
print("Numpy array từ DataFrame:\n", numpy_h)
```
